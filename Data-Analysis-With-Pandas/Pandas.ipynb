# 7) Filter rows of a pandas dataframe by column values

```python
import pandas as pd
movies = pd.read_csv("http://bit.ly/imdbratings")
```

```python
movies.head()
```

```python
movies.shape
```

- Lets say we want to see only movies that are atleast 200 minutes long
- In other words, we want to filter the dataframe so that we still see all the columns but only those rows which have the duration of atleast 200
- python booleans : True and False

### Here's how we will do it step by step
- We will create a python list of booleans that has the same exact length as the dataframe
- Booleans will be **True** if the given row is atleast 200 mins and **False** if its not
- Example for the above shown first 5 rows, our list of booleans will be like [False, False, True, False, False]
- So in other words we are creating a list of booleans that is telling us which rows match a given condition.
- Once we have booleans list, we convert it to a pandas series.
- Pass the series to the dataframe using bracket notation.
- Normally the bracket notation is used to pull out a column from dataframe, but here given a series of booleans, pandas will return the dataframe with all columns but only those rows for which condition was true in the provided series.

```python
booleans = []
for length in movies.duration:
    if length>=200:
        booleans.append(True)
    else:
        booleans.append(False)
```

```python
booleans[0:5]
```

```python
len(booleans) # Same length as the dataframe
```

```python
# convert to pandas series
is_long = pd.Series(booleans)
```

```python
is_long.head()
```

```python
movies[is_long]
```

### The above is the long way of doing so, there is a shorter way

- We can avoid writing for loop, we can directly compare a series using relative/comparision operators and that can return us a new series of booleans.

```python
is_long = (movies.duration >= 200)
# The above condition will directly return a series of booleans, without any need to write the for loop
```

```python
is_long.head()
```

```python
movies[is_long]
```

### Even shorter way
- With just a single line

```python
movies[movies.duration>=200]
```

### Bonus Tip
- Lets say we are just concerned about the genre of the movies that are atleast 200 mins in duration
- Since the above gives us a dataframe, so we can select a series from that dataframe either using dot or bracket notation

```python
movies[movies.duration>=200]['genre']
```

### Note : The above way (i.e. movies[movies.duration>=200]['genre']) will usually work fine but sometimes it will cause strange behaviour
- The better practice now is to use the **loc()** method of the dataframe
- loc() method allows you to select rows and columns by label

```python
movies.loc[movies.duration>=200, 'genre']
```

# 8) Apply multiple filter criteria to a pandas dataframe

```python
import pandas as pd
movies = pd.read_csv('http://bit.ly/imdbratings')
print (movies.head())
```

- Lets look at only long (>=200 min) movies and of genre 'Drama'
- We will use logical operators ('and', 'or' and 'not')

```python
True or False
```

```python
True and False
```

```python
not False
```

```python
# So conceptually what we want to do is to check multiple conditions
# However the below code does not work.
movies[movies.duration>=200 and movies.genre=='Drama']
```

- **We need to add in some parenthesis to help pandas understand the evaluation order. It needs to understand in what order to evaluate these separate statements.**
- **Instead of 'and', use '&' operator**

```python
True & False
```

```python
movies[(movies.duration>=200) & (movies.genre=='Drama')]
```

- To see all movies that are either longer than 200 min OR with genre 'Drama', we use the **'|'** operator

```python
movies[(movies.duration>=200) | (movies.genre=='Drama')]
```

- So in summary, when we need to filter using multiple criterias, we have to put all the conditions in separate parenthesis and separate them with either a '&' for AND or '|' for OR

### Bonus tip:
- What if we have a bunch of OR conditions 
- Lets say we need to look for movies with genre as either 'Crime' or 'Drama' or 'Action'

```python
# Normally we would think of doing it as below
movies[(movies.genre=='Crime') | (movies.genre=='Drama') | (movies.genre=='Action')]
```

- There is a neat trick to simplify multiple OR conditions like this
- We will use **isin()** method of a series
- We pass to this method, a list of all values 
- This method generates a series of Booleans
- We can pass that series to the dataframe using bracket notation

```python
movies.genre.isin(['Crime', 'Drama', 'Action'])
```

```python
movies[movies.genre.isin(['Crime', 'Drama', 'Action'])]
```

# 13) When to use 'groupby' in pandas

```python
import pandas as pd
drinks = pd.read_csv('http://bit.ly/drinksbycountry')
drinks.head()
```

- Lets say our problem is to find the 'average beer servings across all countries'
- We can do it simply as below

```python
drinks.beer_servings.mean()
```

- What if we wanted to know 'average beer serving across all continents'
- Here we use **groupby()** method of dataframe
- Just add a groupby('continent') as below

```python
drinks.groupby('continent').beer_servings.mean()
```

- To see how it works, lets filter the database by 'Africa'

```python
drinks[drinks.continent=='Africa']
```

- Now to calculate mean of this data we just do as usual

```python
drinks[drinks['continent']=='Africa'].beer_servings.mean()
```

- This is what groupby() is doing
- It is filtering the database by 'continent' 
- And then it takes the mean multiple times for each value of continent

### Groupby() is useful anytime we have to analyse some pandas series by some category.
- So in our case category was 'continent' and we had to analyse beer_servings by continent
- So if we can rephrase our question as a "for each", then that's a good candidate for groupby()
- eg. for each continent, what is the mean beer servings

```python
# Max beer serving by continent
drinks.groupby('continent').beer_servings.max()
```

```python
# Min beer serving by continent
drinks.groupby('continent').beer_servings.min()
```

- Instead of using multiple aggregation functions (like mean, max, min) separately, we can use the method **agg()** and specify multiple aggregation functions at once

```python
drinks.groupby('continent').beer_servings.agg(['count', 'min', 'max', 'mean'])
```

### Bonus Tip 1 :
- If we dont supply the column name after groupby(), it will aggregate along all of the numeric columns

```python
drinks.groupby('continent').mean()
```

### Bonus Tip 2:
- We can easily display this in visual form

```python
# This allows plot to appear in the notebook
%matplotlib inline
```

```python
# This shows the information in a side-by-side bar plot
drinks.groupby('continent').mean().plot(kind='bar')
```

# 15) Handling missing values in pandas

```python
import pandas as pd
ufo = pd.read_csv("http://bit.ly/uforeports")
```

```python
ufo.tail()
```

- **NaN** stands for '**Not a Number**', but conceptually it means a missing value.
- It means that when building a dataframe, read_csv() detected a missing value and flagged it as a **NaN**

- To work with missing values, lets look at some methods
- First method is **isnull()**. Its a dataframe method and returns a dataframe.
- **isnull()** shows true if something is missing, otherwise false.

```python
ufo.isnull().tail()
```

- There is also an inverse method of this, i.e. method **notnull()**.

```python
ufo.notnull().tail()
```

- We can use a **sum()** method on returned dataframe of **isnull()** method, it will give a count of null values in each column.

- We can also select this using row position and column label, as below

```python
drinks.ix[1, 'beer_servings']
```

- **ix** has to figure out which is a label and which is a position
- **Note - ix** when used with range has a little different logic
- The logic is that if we have a string labels (like 'beer_servings','spirit_servings') and we use numbers, those numbers are treated as positions. **Positions, as in iloc, are exclusive of end of range.**
- But string labels used as they are, will be inclusive of the end of range.
- Eg. as shown below, the columns are string labels, but we use integers in **ix**, so in the output, it shows only columns in positions [0,2). But for rows in includes range ['Albania', 'Andorra']

```python
drinks.ix['Albania':'Andorra', 0:2]
```

**Note**
- If we have integer index, and we it numbers to **ix**, it will treat it as labels, and will be inclusive of both start and end of the range
- Eg. as below the dataframe returned has 3 rows, but only 2 columns. This is because 'ufo' dataframe had integers as index and string as column names. Therefore the column names when used with position values, the end of range is excluded. But for rows the end of row is included.

```python
ufo.ix[0:2, 0:2]
```

- Always prefer using **loc** and **iloc**, over **ix**. That is the recommended method.
- With **ix**, you have to keep in mind the intricate logic of mixing labels and integers. Also it gets confusing when integer labels don't start with 0 or are not continuous.

# 18) When to use 'inplace' parameter in pandas

```python
import pandas as pd
```

```python
ufo = pd.read_csv('http://bit.ly/uforeports')
ufo.shape
```

```python
ufo.head()
```

- Let's say we dont care about the 'City' column so we drop it from the dataframe.

```python
ufo.drop('City', axis=1)
```

- The 'City' column is gone above, but its not actually gone from the original dataframe.

```python
ufo.head()
```

- It's not actually dropped from the dataframe, because the **drop()** method has an **inplace** parameter that is **by default set to False**.
- Any operation with an inplace parameter set to False, means that the operation will not affect the underlying data by default.
- To make it permanent, set **inplace=True**

```python
ufo.drop('City', axis=1, inplace=True)
```

- On running the above **drop()** method with **inplace=True**, we see that nothing prints out. This was not the case with **inplace=False**, which printed the new dataframe with no 'City' column.
- This is usually a tip off in python, that something has happened inplace.

```python
ufo.head()
```

- The inplace parameter is by default set to False for all the methods that support it. This gives us the option to experiment with data, check the impact of an operation on data without affecting the data permanently, but if required, proactively set inplace=True to modify the dataframe.

- Assignment using '=' can also give the same result as inplace=True.
- Consider the below line

```python
ufo = ufo.set_index('Time')
# will give the same result as ufo.set_index('Time', inplace=True)
```

```python
ufo.tail()
```

- The assignment operator might create a separate copy, so essentially there would be two copies of the dataframe temporarily. So if the dataframe is large, running into GB's, then it will affect the system performance.
- There is **no gaurantee**, that with inplace=True, the system performance will be better, however it is recommended to use inplace=True over assignment statement, as it may be better.

### Bonus Tip
- How to take advantage of inplace=False default to explore pandas methods??
- It can help figure out different parameter settings for some method while experimenting with data. (i.e. parameters other than inplace)
- Eg. **fillna()** method of the dataframe has various methods to fill via **method** parameter. (see all the options with shift+tab). To try out different options, as to check what works for you, you can experiment with inplace=False, as it will not affect the underlying dataframe.

```python
ufo.fillna(method='bfill').tail()
```

```python
ufo.fillna(method='ffill').tail()
```

# 19) How to make pandas Dataframe smaller and faster

```python
import pandas as pd
```

```python
drinks = pd.read_csv('http://bit.ly/drinksbycountry')
drinks.head()
```

- We can use the **info()** dataframe method to get more info on the dataframe. It can give information about column types and total memory occupied etc.

```python
drinks.info()
```

- type 'object' mostly means string type, but we can also create a pandas Series of python lists or a pandas Series of python dictionaries.
- We can store arbitrary python objects in the pandas Series, and pandas will basically just store a reference to that object, and calls it type 'object'.
- The memory usage at the bottome says '7.6+ KB'
- It tells us that the dataframe takes atleast 7.6 KB of memory. The '+' is there because the 'object' columns are references to other objects and pandas wants to run the **info()** method faster, so it doesn't go out and look at the referenced objects to figure out how much space they take. It only reports how much space the references take, not the actual referenced object.

- It turns out that we can force pandas to count the true memory usage.
- Use parameter **memory_usage='deep'** in **info()** method

```python
drinks.info(memory_usage='deep')
```

- To check how much space each column/series is taking we can use the dataframe method **memory_usage()**.
- It gives the memory usage of each column in bytes.
- Similar to **info()** method, it does not inspect the 'object' columns by default. Use parameter **deep=True** to get the actual space occupied.

```python
drinks.memory_usage()
```

```python
drinks.memory_usage(deep=True)
```

```python
# Since the method returns a Series, we can call the sum() method and get a value similar to 
# total bytes reported by info() method
drinks.memory_usage(deep=True).sum()
```

- Bottom line is that the 'object' columns can take up a lot of space.
- So question is how to be more space efficient, especially with 'object' columns.
- What if we were able to store our strings as integers, because integers are more space efficient than strings.
- Eg. There are only 6 unique values in the 'Continent' Series in our dataframe. However it takes a lot of space.

```python
sorted(drinks.continent.unique())
```

- We can store it in integers by mapping integer 0 to 'Africa', a 1 to 'Asia', and so on.
- By doing that we are only storing integers.
- So we have to store a lookup table to perform the mapping, but the strings are stored just once in the lookup table, instead of being stored multiple times in the dataframe column.
- Pandas has created this system for us, we don't have to implement all by scratch.
- There is a **category datatype**, which can be used to convert 'object' column to more space efficient way.

```python
drinks['continent'] = drinks.continent.astype('category')
```

```python
drinks.dtypes
```

- Now the datatype of 'continent' column is not 'object' but 'category'
- But it still looks the same when printed, because it gets the actual string from the lookup table

```python
drinks.head()
```

```python
drinks.continent.head()
```

- As seen above, the type is 'category'
- Also it lists the categories **Categories (6, object): [Africa, Asia, Europe, North America, Oceania, South America]**
- Under the hood, it is storing the strings as integers

- As we had seen with string objects/Series, we can use string methods by calling like **drinks.continent.str.something()**
- Similarly with categorical Series, we use **drinks.continent.cat.something()**

```python
drinks.continent.cat.codes.head()
```

- As shown in above output, this is exactly how pandas is now representing the continent series as integers.
- Lets take a look at the memory usage now.

```python
drinks.memory_usage(deep=True)
```

- **The memory usage of 'continent' series has dropped from 6928 bytes earlier to 416 bytes.**
- Instead of storing 193 strings, we are now storing 193 integers that point to a lookup table of 6 strings.
- So strings have to be stored just once, the rest is integers storage, thus much more space effficient.

```python
drinks['country'] = drinks.country.astype('category')
```

```python
drinks.memory_usage(deep=True)
```

- This didn't go very well, rather the memory usage has increased.
- This is because the country names were unique, and we have used up extra data for 193 integer storage and lookup table.
- The bottomline is that to use the 'category' datatype when you have an 'object' column of strings that have only a few different values.

- Besides just reducing the memory usage, it **also speeds up the the computations.**
- If we are working with an object column of strings and we are doing computations with that column (eg. groupby), it will actually speed up the operations if we use category datatype.

# 25) Avoid a SettingWithCopyWarning in pandas
- Let us look at scenarios that generate this warning

```python
import pandas as pd
```

```python
movies = pd.read_csv('http://bit.ly/imdbratings')
movies.head()
```

```python
# Count number of missing values for content_rating
movies.content_rating.isnull().sum()
```

```python
movies[movies.content_rating.isnull()]
```

```python
# Lets look at the values in content_rating
movies.content_rating.value_counts()
```

- Lets decide for this lecture, that the content_rating 'NOT RATED' should be represented as a missing value.
- So we replace them with NaN, so that we can take advantage of missing value functionality. NaN comes from numpy so we import numpy.

```python
import numpy as np
```

```python
# Overwrite the 'NOT RATED' values with np.nan
movies[movies.content_rating=='NOT RATED'].content_rating = np.nan
```

- And we get the **SettingWithCopyWarning**.
- First, lets take a look and see if it worked, because it is just a warning and not an error.

```python
movies.content_rating.isnull().sum()
```

- So we see that it didn't work. It should have been 68.
- The solution is mentioned in the warning, and is as below.
- Use the loc method.

```python
movies.loc[movies.content_rating=='NOT RATED', 'content_rating'] = np.nan
```

```python
movies.content_rating.isnull().sum()
```

- So it did work.
- Reason for not working :-
    - **movies[movies.content_rating=='NOT RATED'].content_rating = np.nan**
        - This is actually two operations
        - movies[movies.content_rating=='NOT RATED'] --> this part is a 'get item' operation
        - .content_rating = np.nan --> this part is a 'set item' operation with a reference to whatever was produced by the first part.
        - So we have a 'get item' and a 'set item'
        - Pandas can't gaurantee whether the first part (i.e. 'get item') returned a view or a copy of the data.
        - If it returned a view of the data, the second part (i.e. 'set item') will affect the dataframe.
        - But if it returned a copy, it would be modifying the copy and the original dataframe is not modified.
        - So pandas is trying to warn that it does not know whether it has a view or a copy of the data.
    - **movies.loc[movies.content_rating=='NOT RATED', 'content_rating'] = np.nan**
        - This is a single 'set item' operation
        - That's why it does not throw any warning.
        
- **Summary here is that if you are trying to select rows and columns in the same line of code, use .loc and that will work as expected.**

- Lets see a second example of the warning
- We will pretend here that we only want to focus on movies with very high star ratings.

```python
top_movies = movies.loc[movies.star_rating >= 9, :]
```

```python
# We have only 4 such movies with start rating >= 9
top_movies
```

- Suppose we notice that the duration for movie 'The Shawshank Redemption' is wrong and we want to fix it.

```python
top_movies.loc[0, 'duration'] = 150
```

- Again we get the same warning, **even though we used .loc**
- And this is confusing, because the suggestion in the warning is to use .loc[], which is what we are using.
- Before understanding the problem, lets check if it works.

```python
top_movies
```

- As we can see, it did work and modified the dataframe.
- This is also a reason to not turn off the warnings, because inspite of warning sometimes they work and sometimes they don't. So it is an indication to check manually whether it has worked or not.
- **Reason for warning in this case :-**
    - pandas was not sure whether **top_movies** is a view or a copy of **movies** dataframe.
    - So it is trying to warn you that whether you are trying to modify one dataframe or are you trying to modify two dataframes.
    - So the problem was due to the line **top_movies = movies.loc[movies.star_rating >= 9, :]**

```python
# it did not modify the movies dataframe, only the top_movies dataframe
movies.head()
```

- The **solution** is that, anytime you are trying to create a dataframe copy, you should explicitly use the dataframe's **copy()** method, and then pandas can be sure that it is a copy and won't give any warning.

```python
top_movies = movies.loc[movies.star_rating >= 9, :].copy()
```

```python
top_movies.loc[0, 'duration'] = 150
```

```python
# No warnings and it modified the dataset
top_movies
```

- Lets use apply() to solve a harder problem.
- Lets extract the last name of each person into its own column.
- We need to get the part before the ',' int the 'Name' column.
- We can split it using the str.split() method on the series.

```python
train.Name.str.split(',').head()
# This gives a Series of python lists. Each entry has two strings.
```

- But how to just get the first part.
- We have got a series of lists, and we just want the first element of it. So we apply a function on this series.
- What we really need to do is :-
    - write a function to extract the first part of the list.
    - apply that function using apply() method on this series of list.

```python
def get_element(my_list, position):
    return my_list[position]
```

```python
train.Name.str.split(',').apply(get_element, position=0).head()
```

- This tells pandas to get the series **'train.Name.str.split(',')'**, then apply function **get_element** on every element of series and pass keyword argument **position=0** to the function.
- **Note - **We can also do this using lambda functions as below

```python
train.Name.str.split(',').apply(lambda x: x[0]).head()
```

- Now lets see apply() as a dataframe method.
- It applies a function along either axis of the dataframe

```python
drinks = pd.read_csv('http://bit.ly/drinksbycountry')
```

```python
drinks.head()
```

```python
# We are working with a subset of the dataframe
drinks.loc[:, 'beer_servings':'wine_servings']
```

- We want to find the maximum value in each type of serving.
- That is, we move along the rows and take the maximum (apply python max() function)

```python
drinks.loc[:, 'beer_servings':'wine_servings'].apply(max, axis=0)
```

- Now we want to find the max among the 3 servings in each row.
- That is, we move in the direction of columns and apply the max() function

```python
drinks.loc[:, 'beer_servings':'wine_servings'].apply(max, axis=1)
```

- If you want to know which column is the maximum, instead of knowing the maximum value.
- Use **np.argmax()** method to find that.

```python
# This tells which row had the maximum value for each kind serving
drinks.loc[:, 'beer_servings':'wine_servings'].apply(np.argmax, axis=0)
```

```python
# This tells the kind of serving in each row which is maximum
drinks.loc[:, 'beer_servings':'wine_servings'].apply(np.argmax, axis=1)
```
